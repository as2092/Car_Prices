```python
#deleting the rows where the target variable "price" has missing values
numeric_cars= numeric_cars.dropna(subset = ['price'])
numeric_cars.isnull().sum()
```

```python
# Filling the null values with the mean of the columns as every column is continuous variable
numeric_cars.fillna(numeric_cars.mean(), inplace=True)
```

```python
# Normalize all columnns to range from 0 to 1 except the target column.
price_col = numeric_cars['price']
numeric_cars = (numeric_cars - numeric_cars.min())/(numeric_cars.max() - numeric_cars.min())
numeric_cars['price'] = price_col
numeric_cars
```

```python
from sklearn.model_selection import cross_val_score, KFold
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error
```

```python
def knn_train_test(df, features, target_col):
    np.random.seed(1)
        
    shuffled_index = np.random.permutation(df.index)
    sh_df = df.reindex(shuffled_index)

    mid = int(len(sh_df) / 2)
    train_df = sh_df.iloc[0: mid]
    test_df = sh_df.iloc[mid: ]
    
    knn = KNeighborsRegressor()
    
    knn.fit(train_df[[features]], train_df[target_col])
    predictions = knn.predict(test_df[[features]])
    
    mse = mean_squared_error(test_df[target_col], predictions)
    rmse = np.sqrt(mse)
    
    return rmse  
```

## Multivariate Model

```python
def knn_train_test2(df, features, target_col, k):
    np.random.seed(1)
        
    shuffled_index = np.random.permutation(df.index)
    sh_df = df.reindex(shuffled_index)

    mid = int(len(sh_df) / 2)
    train_df = sh_df.iloc[0: mid]
    test_df = sh_df.iloc[mid: ]
    
    knn = KNeighborsRegressor(n_neighbors=k)
    
    knn.fit(train_df[features], train_df[target_col])
    predictions = knn.predict(test_df[features])
    
    mse = mean_squared_error(test_df[target_col], predictions)
    rmse = np.sqrt(mse)
    
    return rmse  
```
